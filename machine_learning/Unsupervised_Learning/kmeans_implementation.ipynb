```python
def cal_distance(X1, X2):
    return (sum((X1 - X2)**2))**0.5
```

```python
def assign_cluster(centroids, cluster_data):
    
    clusters = []
    
    for i in range(cluster_data.shape[0]): 
        distance = []
        for centroid in centroids:
            dis = cal_distance(centroid, cluster_data.iloc[i])
            distance.append(dis)
            
        cluster = [cls for cls,val in enumerate(distance) if val==min(distance)]
        clusters.append(cluster[0])
        
    return clusters
```

```python
import pandas as pd
```

```python
def cal_centroids(cluster, cluster_data):
    
    new_centroids = []
    
    cluster_df = pd.DataFrame(cluster, columns=['cluster'])
    cluster_data_df = pd.DataFrame(cluster_data)
    
    df = pd.concat([cluster_data_df, cluster_df], axis=1)
    
    clusters = set(df['cluster'])
    
    for cluster in clusters:
        
        cols = df.columns[:-1]
        df_ = df[df['cluster']==cluster][cols]
        cluster_mean = df_.mean(axis=0)
        
        new_centroids.append(cluster_mean)
        
    return new_centroids   
```
